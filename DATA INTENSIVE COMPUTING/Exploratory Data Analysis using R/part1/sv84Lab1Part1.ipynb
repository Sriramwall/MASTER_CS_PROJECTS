<h1>Problem 1</h1>

```r
sales1<-c(12,14,16,29,30,45,19,20,16, 19, 34, 20)
for(data in c(20,30,40)){
    sales2<-rpois(data,data+20)  # random numbers, Poisson distribution, mean at 34, 12 numbers
    par(bg="cornsilk")

    plot(sales1, col="blue", type="o", ylim=c(0,100), xlab="Month", ylab="Sales" )
    title(main="Sales by Month")

    lines(sales2, type="o", pch=22, lty=2, col="red")
    grid(nx=NA, ny=NULL)
    legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
}
```

<h3>Problem 1) DISCUSSION:</h3>

In the above problem Sales1 vector is plotted by giving the values manually and thus it remains the same in all the three images.
<br>
<br>
Sales2 vector is plotted by calling rpois function in a loop which produces the randomness in all the three images.
Thus the randomness of the second set of data has been observed.


<h1>Problem 2</h1>

```r
sales<-read.table("salesdata.txt", header=T)
sales  # to verify that data has been read
barplot(as.matrix(sales), main="Sales Data", ylab= "Total",beside=T, col=rainbow(5))
```

<h3>Problem 2 DISCUSSION</h3>
A histogram is the most commonly used graph to show frequency distributions. A frequency distribution shows how often each different value in a set of data occurs.
<br>
<br>
As you can see from the graph, it is easy to analyse which Sales data has produced maximum sales value(Sales 2 in this case) and which has produced identical sales value in consecutive stages(Sales 1 in this case).
<br>
<br>
Thus a histogram has been plotted which shows the frequency of score occurrences in a continuous data set that has been divided into classes, called bins which are Sales 1 and Sales 2 in this case.


<h1> Problem 3 </h1>

```r
fn<-boxplot(sales,col=c("orange","green"))$stats
fn #values of all quartiles, list -> one coloumn of data followed by another
text(1.45, fn[3,2], paste("Median =", fn[3,2]), adj=0, cex=.7)
text(0.45, fn[3,1],paste("Median =", fn[3,1]), adj=0, cex=.7)
grid(nx=NA, ny=NULL)
```

<h3>Problem 3 DISCUSSION</h3>
This type of graph is used to show the shape of the distribution, its central value, and its variability. The ends of the box are the upper and lower quartiles, so the box spans the interquartile range. The median is marked by a vertical line inside the box.
<br>
<br>
In the above problem, we can see the interquartile range of sales 1 and sales 2 and their corresponding medians which are at 34 and 45 respectively. The upper quartile, lower quartile, whiskers of the 2 sales data can also be seen from the graph.
